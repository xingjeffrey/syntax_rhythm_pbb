```python
tempo_min = min(dyadic['cycle_dur'].values)
tempo_max = max(dyadic['cycle_dur'].values)
tempo_bins = np.linspace(tempo_min, tempo_max, 101)
```

```python
ratio_min = 0
ratio_max = 1
ratio_bins = np.linspace(ratio_min, ratio_max, 101)
```

```python
tempo_bins
```

```python
ratio_bins
```

```python
tem = []
cnt = []
rat = []

for i in np.arange(0, 100):
    counter = dyadic.copy(deep = True)
    ## find tempo range
    #print(tempo_bins[i], tempo_bins[i+1])
    if tempo_bins[i+1] > max(tempo_bins):
        break
    tempo = counter.loc[(counter['cycle_dur']>=tempo_bins[i])&(counter['cycle_dur']<tempo_bins[i+1])]
    
    ## find ratio range
    for j in np.arange(0, 100):
        #print(ratio_bins[i], ratio_bins[i+1])
        ratio = tempo.loc[(tempo['ratio_roeske']>=ratio_bins[j])&(tempo['ratio_roeske']<ratio_bins[j+1])]
        
        tem.append(tempo_bins[i])
        rat.append(ratio_bins[j])
        cnt.append(len(ratio))
        
dyadic_bins = pd.DataFrame(
    {
        'tempo':tem,
        'ratio':rat,
        'cnt':cnt
    }
).reset_index()

dyadic_bins
```

```python
from sklearn.neighbors import KernelDensity
import scipy.signal
import numpy as np
```

```python
normalized_dyadic_bins = []

## for each tempo bin
for tempo in np.unique(dyadic_bins['tempo'].values):
    tempo_slice = dyadic_bins[dyadic_bins['tempo'] == tempo]
    cnt_sum = sum(tempo_slice.cnt.values)
    ## prevent 0 division
    if cnt_sum == 0:
        cnt_sum = 1
        
    ## get a probability density
    tempo_slice['prob_density'] = tempo_slice['cnt'] / cnt_sum
    
    ## if the tempo slice has less than 10 rhythm ratios, ignore it
    if cnt_sum < 15:
        tempo_slice['prob_density'] = 0
    
    ## get peakiness
    tempo_slice['peak'] = 0
    peaks = scipy.signal.find_peaks(tempo_slice['cnt'], 0.07)[0]
    for peak in peaks:
        tempo_slice.loc[tempo_slice['ratio'] == tempo_slice['ratio'].values[peak], 'peak'] = 1
            
        
    normalized_dyadic_bins.append(tempo_slice)
    
normalized_dyadic_bins = pd.concat(normalized_dyadic_bins)
```

```python
normalized_dyadic_bins['tempo'] = np.round(normalized_dyadic_bins['tempo']*1000, 0)
normalized_dyadic_bins['ratio'] = np.round(normalized_dyadic_bins['ratio'], 2)
```

```python
normalized_dyadic_bins['grand_prob_density'] = (normalized_dyadic_bins['cnt'] - np.min(normalized_dyadic_bins['cnt']))/(np.max(normalized_dyadic_bins['cnt']-np.min(normalized_dyadic_bins['cnt'])))
```

```python
norm_dyadic_bins_pivot = normalized_dyadic_bins.pivot('tempo', 'ratio', 'grand_prob_density')

figure(figsize = (8, 6), dpi = 300)

plt.rcParams['patch.edgecolor'] = 'none'

fig = sns.heatmap(
    data = norm_dyadic_bins_pivot
)

figname = 'rhythm_tempo' + indv + '.pdf'
plt.savefig(FIGURE_DIR/'categorical_rhythm'/figname)
```
