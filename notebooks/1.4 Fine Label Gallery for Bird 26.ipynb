## Load in Data

```python
import pandas as pd
```

```python
note_df = pd.read_pickle(SAVE_LOC)
```

```python
indv_df = note_df[note_df['indv'] == indv].reset_index()
```

```python
indv_df
```

## Graph spectrographic mean of each label

```python
import numpy as np
import matplotlib.pyplot as plt
from librosa.util import pad_center
from joblib import Parallel, delayed
from butcherbird.utils.paths import FIGURE_DIR
```

```python
unique_labels = np.unique(note_df['lasso_labels'])
unique_labels
```

```python
spec_means = []
spec_n = []

## for each unique label
for label in unique_labels:
    ## find spectrographic mean
    specs = indv_df[indv_df['lasso_labels'] == label]['spectrogram'].values
    
    spec_n.append(len(specs))
    
    spec_lens = [np.shape(i)[1] for i in specs]
    pad_length = np.max(spec_lens)
    
    with Parallel(n_jobs=36, verbose=0) as parallel:
                specs = parallel(
                    delayed(pad_center)(spec, pad_length)
                    for spec in specs
                )
    spec_means.append(np.average(np.array(specs), axis = 0))
```

```python
spec_means_lens = [np.shape(i)[1] for i in spec_means]
max_length = np.max(spec_means_lens)
```

```python
font = {'family' : 'normal',
        'size'   : 10}
plt.rc('font', **font)
```

```python
for label, n, spec_mean in zip(unique_labels, spec_n, spec_means):
    
    fig, ax = plt.subplots(figsize=((np.shape(spec_mean)[1] / max_length) * 15, 3))

    ax.set_title('Label ' + str(label) + ' (n = ' + str(n) + ')')
    ax.imshow(spec_mean, origin='lower')
    ax.set_ylim([0, np.shape(spec_mean)[0]])
    ax.set_xlim([0, np.shape(spec_mean)[1]])
    plt.xticks(np.arange(0, np.shape(spec_mean)[1], 50))
    plt.yticks(np.arange(0, np.shape(spec_mean)[0] + 1, 64))
    
    labels = [item.get_text() for item in ax.get_yticklabels()]
    labels[1] = 3
    labels[0] = 0.5
    
    ax.set_yticklabels(labels)
    
    
    plt.xlabel('Time (ms)')
    plt.ylabel('Freq. (kHz)')
    fig.set_dpi(300)
    
    plt.tight_layout()
    
    # save
    figname = 'label_gallery_' + indv + '_label_' + str(label) + '.pdf'
    plt.savefig(FIGURE_DIR/'label_gallery'/figname, pad_inches=1)
```
