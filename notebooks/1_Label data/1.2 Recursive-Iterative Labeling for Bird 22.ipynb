## Load up Note Dataframe

## Visualize Coarse Labels

```python
from butcherbird.visualization.projections import scatter_spec
import matplotlib.pyplot as plt
import numpy as np
```

```python
scatter_spec(
    np.vstack(note_df.umap),
    note_df.spectrogram, 
    column_size = 8,
    pal_color = 'tab20',
    color_points = False,
    enlarge_points = 3, 
    figsize = (8,8),
    scatter_kwargs = {
        'labels': list(note_df['hdbscan_labels'].values),
        'alpha':0.75, 
        's': 0.25,
        'show_legend': True
    },
    matshow_kwargs = {
        'cmap': plt.cm.Greys
    },
    line_kwargs = {
        'lw':1,
        'ls':'dashed',
        'alpha':0.25,
    },
    draw_lines=True,
    n_subset = 1000,
    border_line_width = 3,
);
plt.show()
```

## Discover Subclusters via Recursive UMAP

```python
from butcherbird.label.label import labeler, recursive_umap
from tqdm.autonotebook import tqdm
from datetime import datetime
```

```python
pd.options.mode.chained_assignment = None  # default='warn', ignore warning for right now
```

```python
datetime.now().isoformat(timespec='minutes')
```

```python
note_df = recursive_umap(
    note_df, 
    log_scaling_boolean = False,           
    log_scaling_factor = 10,
    lin_scaling_boolean = True,           
    lin_scaling_factor = 0.4,
    n_jobs = 24,                                                              
    min_dist = 0,                           
    n_neighbors_factor = 0.1,                                                                
    stopping_nn = int(0.002 * len(note_df)),                                                           
    min_cluster_size_factor = 0.1,                                                           
    stopping_cluster_size = int(0.002 * len(note_df))
)
```

```python
datetime.now().isoformat(timespec='minutes')
```

## Clean Data

```python
from butcherbird.label.label import clean_recursive_umap
```

```python
note_df = clean_recursive_umap(note_df)
```
