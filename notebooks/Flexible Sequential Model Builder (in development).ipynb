```python
import numpy as np
```

```python
note_df_dc = note_df.copy(deep = True)
```

```python
seq = list(note_df_dc['lasso_labels'].values)
unique_states, counts_states = np.unique(note_df_dc['lasso_labels'].values, return_counts = True)
p_states = counts_states/len(note_df_dc['lasso_labels'].values)
```

```python
order = 2
```

```python
## Make empty object
transition_object = np.zeros(np.full(order + 1, fill_value = len(unique_states)))
```

```python
transition_object
```

```python
## for every current index that has adequete amount of history
for i in np.arange(len(seq[order:]) - order):
    
    ## the amount of history we need to go back is equivalent to order
    order_ctr = order

    ## initialize indexer to tell transition object where to tally
    indexer = []
    
    while order_ctr >= 0:
        indexer.append(seq[i])
        order_ctr = order_ctr - 1
        i = i + 1
        
    print('tally at',indexer)
    ## tally at destination
    transition_object[tuple(indexer)] = transition_object[tuple(indexer)] + 1
    print(transition_object[tuple(indexer)])
```
